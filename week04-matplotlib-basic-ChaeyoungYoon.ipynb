```python
import matplotlib.pyplot as plt
import numpy as np 

x= np.linspace(0, 2*np.pi ,200)
y=np.sin(x)
plt.plot(x,y)

plt.grid(True)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False

plt.title('삼각함수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

x= np.linspace(0, 2*np.pi ,200)
y=np.sin(x)
plt.plot(x,y)

plt.grid(True,ls=':')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] =False

plt.title('삼각함수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N=8
y=np.zeros(N)

x1=np.linspace(0,10,N,endpoint=True)
x2=np.linspace(0,10,N,endpoint=False)
plt.plot(x1,y+0.5,'o', label='endpoint=True')
plt.plot(x2,y,'o', label='endpoint=False')

plt.legend()
plt.ylim([-0.5,1])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N=8
y=np.zeros(N)

x1=np.linspace(0,10,N,endpoint=True)
x2=np.linspace(0,10,N,endpoint=False)
plt.plot(x1,y+0.5,'o-', label='endpoint=True')
plt.plot(x2,y,'o:', label='endpoint=False')

plt.legend()
plt.ylim([-0.5,1])
plt.show()
```
